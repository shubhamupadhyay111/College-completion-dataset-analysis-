#**COLLAGE COMPLETION DATASET ANALYSIS**

### Dataset Introduction

This dataset contains detailed information about 3,798 colleges across the United States, with 49 attributes describing various aspects of each institution. Key features of the dataset include:

1. **General Information:**
   - College name, city, and state.
   - College level (e.g., 4-year or 2-year).
   - Type of institution (public, private, etc.).

2. **Student and Institutional Statistics:**
   - Total student count.
   - Percentage of full-time students.
   - On-time and late graduation rates.

3. **Financial Data:**
   - Average financial aid provided.
   - Endowment values.
   - Spending on awards.

4. **Performance Metrics:**
   - Median SAT scores and percentiles.
   - Retention rates.
   - Graduation rates (standard and transfer students).

5. **Specialized Metrics:**
   - Number of awards per 100 students and per state.
   - Faculty metrics, such as full-time faculty percentages.
   - Enrollment and graduation data for transfer students.

Some columns have missing values, such as median SAT scores, endowment data, and transfer graduation rates, which may require attention during analysis. The dataset provides a rich resource for exploring trends in higher education, assessing institutional performance, and making comparative analyses between colleges.

#1. What is the average graduation rate (on_time_graduate) across all colleges in the dataset?


```python
# @title Average Graduation Rate Across All Colleges
# Calculate the average graduation rate
average_graduation_rate = df['on_time_graduate'].mean()

# Create a bar chart
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=['Average'], y=[average_graduation_rate],color='orange')
plt.title('Average Graduation Rate Across All Colleges')
plt.ylabel('Graduation Rate')
plt.ylim(0, 100)

# Add data labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

This bar chart displays the average on-time graduation rate across all colleges in the dataset. The average graduation rate is around 27.85%.

#2. Is there a correlation between the student count (student_count) and the on-time graduation rate?

```python
# @title Correlation between Student Count and On-Time Graduation Rate
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.scatterplot(x='student_count', y='on_time_graduate', data=df,color='red')
plt.title('Correlation between Student Count and On-Time Graduation Rate')
plt.xlabel('Student Count')
plt.ylabel('On-Time Graduation Rate')
plt.show()
```

 The scatter plot shows the relationship between the number of students in a college and its on-time graduation rate. There's a slight negative correlation, implying colleges with fewer students tend to have a marginally higher on-time graduation rate. However, the correlation isn't very strong

#3. How does the average financial aid (aid_value) received by students vary across different states (state)?

```python
# @title Average Financial Aid Received by Students Across Different States
state_aid = df.groupby('state')['aid_value'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='aid_value', data=state_aid,color='purple')
plt.title('Average Financial Aid Received by Students Across Different States')
plt.xlabel('State')
plt.ylabel('Average Financial Aid')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

The bar chart compares the average financial aid received by students in different states. The chart highlights variations in financial aid across states.

#4. What is the distribution of expenditure on awards (spent_award_value) for colleges?

```python
# @title Distribution of Expenditure on Awards for Colleges
plt.figure(figsize=(6, 6))  # Adjust figure size if needed
sns.histplot(df['spent_award_value'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Expenditure on Awards for Colleges')
plt.xlabel('Expenditure on Awards')
plt.ylabel('Frequency')
plt.show()
```

The histogram illustrates the distribution of expenditure on awards for colleges. It provides insights into how colleges allocate funds for awards, indicating frequency and range of spending

#5. Is there a relationship between the percentage of full-time students (fulltime_student_per) and the on-time graduation rate?

```python
# @title Relationship between Percentage of Full-Time Students and On-Time Graduation Rate
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.scatterplot(x='fulltime_student_per', y='on_time_graduate', data=df, color='green')
plt.title('Relationship between Percentage of Full-Time Students and On-Time Graduation Rate')
plt.xlabel('Percentage of Full-Time Students')
plt.ylabel('On-Time Graduation Rate')
plt.show()
```

The scatter plot examines the relationship between the percentage of full-time students and the on-time graduation rate. A slight positive correlation can be observed, suggesting colleges with a higher percentage of full-time students tend to have a slightly higher graduation rate

#6. Which state (state) has the highest average awards per 100 students (awards_per_100_value)?

```python
# @title Average Awards per 100 Students by State
state_awards = df.groupby('state')['awards_per_100_value'].mean().reset_index()
state_awards = state_awards.sort_values('awards_per_100_value', ascending=False)
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
sns.barplot(x='state', y='awards_per_100_value', data=state_awards, color='blue')
plt.title('Average Awards per 100 Students by State')
plt.xlabel('State')
plt.ylabel('Average Awards per 100 Students')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

The bar chart compares the average awards per 100 students for different states. It reveals which states have higher average awards per 100 students, highlighting potential variations in award allocation across states.

#7. How does the type of college control (control) (public, private, etc.) influence the on-time graduation rate?

```python
# @title On-Time Graduation Rate by College Control Type
control_grad_rate = df.groupby('colllage_type')['on_time_graduate'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='colllage_type', y='on_time_graduate', data=control_grad_rate, palette=['red','green','blue'])
plt.title('On-Time Graduation Rate by College Control Type')
plt.xlabel('College Control Type')
plt.ylabel('On-Time Graduation Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

This bar chart presents the on-time graduation rates for different college control types (public, private, etc.). It helps compare the graduation rates based on the college's control, potentially revealing differences in graduation outcomes based on ownership or governance structure

#8. Is there a relationship between the college type (colllage_type) and the average retention rate?

```python
# @title Average retention rate by collage type
# Assuming your DataFrame is named 'df' and the retention rate column is named 'retain_value'
average_retention_by_type = df.groupby('colllage_type')['retain_value'].mean().reset_index()

plt.figure(figsize=(8, 6))
ax = sns.barplot(x='colllage_type', y='retain_value', data=average_retention_by_type,palette=['red','orange','yellow'])  # Store the Axes object in 'ax'
plt.title('Average Retention Rate by College Type')
plt.xlabel('College Type')
plt.ylabel('Average Retention Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add data labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

The bar chart displays the average retention rates for different college types. It helps understand how retention rates vary across college types, providing insights into student persistence and success within each type.

#9. What is the average student count (student_count) for colleges in each state?

```python
# @title Average student count for collage in each state
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average student count for each state
avg_student_count_by_state = df.groupby('state')['student_count'].mean().reset_index()

# Create a bar chart with dark pastel colors
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
sns.barplot(x='state', y='student_count', data=avg_student_count_by_state, palette="deep")
plt.title('Average Student Count for Colleges in Each State')
plt.xlabel('State')
plt.ylabel('Average Student Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

This bar chart shows the average student count for colleges in each state. It helps understand the distribution of student populations across states, revealing variations in college sizes.

#10. Which colleges have the highest average awards per 100 students (awards_per_100_value)?

```python
# @title Default title text
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random

college_awards = df.groupby('collage_name')['awards_per_100_value'].mean().reset_index()

top_colleges = college_awards.sort_values('awards_per_100_value', ascending=False).head(10)  # Select top 10 colleges
# Assuming 'top_colleges' DataFrame is already created

# Generate a list of random colors with the desired number of colors
num_colors = len(top_colleges)
colors = [
    '#' + ''.join([random.choice('0123456789ABCDEF') for j in range(6)])
    for i in range(num_colors)
]


# Create the bar chart with the custom color palette
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='collage_name', y='awards_per_100_value', data=top_colleges, palette=colors)

# Add data labels with the same color as the corresponding bar
for p, color in zip(ax.patches, colors):
    ax.annotate(
        f'{p.get_height():.2f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 10),
        textcoords='offset points',
        color=color,
    )

plt.title('Top 10 Colleges with Highest Average Awards per 100 Students')
plt.xlabel('College Name')
plt.ylabel('Average Awards per 100 Students')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

The bar chart highlights the top 10 colleges with the highest average awards per 100 students. It identifies colleges with greater emphasis on award allocation and student support
